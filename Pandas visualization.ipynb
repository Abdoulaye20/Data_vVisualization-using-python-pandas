```python
%matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("california.csv")
data.head()
```

```python
#line chart ploting using pandas 
data.plot.line(x='population',y='median_house_value',figsize=(8,6))
plt.title('population vs median_house_value')
plt.show()
```

```python
#histogram chart plotting using pandas
data.plot.hist(x='population',y='median_house_value',figsize=(8,6))
```

```python
data.plot.box(figsize=(8,6))
plt.show()
```

```python
#ploting of exagonal chart in pandas
data.plot.hexbin(x='population',y='median_house_value',figsize=(8,6))
plt.show()
```

```python
plt.plot(data.population,data.median_house_value)
plt.show()
```

```python
plt.bar(data.population,data.total_rooms)
plt.show()
```

```python
# using karnel density
data['median_house_value'].plot.kde()
plt.show()
```
